---

```python
# calculate the difference between two columns
spread = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
```

```python
# add to the dataframe as a new column, at position 1
clean_df.insert(1, "Spread", spread)
clean_df.head()
```

```python
# sort by a particular column
low_risk = clean_df.sort_values("Spread", ascending=True)
low_risk[["Undergraduate Major", "Spread"]].head()
```

---

```python
# Find the top 5 degrees with the highest values in the 90th percentile. 
```

```python
top_five = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
top_five[["Undergraduate Major", "Mid-Career 90th Percentile Salary"]].head()
```

```python
# Also, find the degrees with the greatest spread in salaries. 
# Which majors have the largest difference between high and low earners after graduation.
spread = clean_df["Mid-Career 90th Percentile Salary"] - clean_df["Mid-Career 10th Percentile Salary"]
high_risk = clean_df.sort_values("Spread", ascending=False)
high_risk[["Undergraduate Major", "Spread"]].head()
```

```python
# compare highest spread vs. highest median salary
high_median = clean_df.sort_values("Mid-Career Median Salary", ascending=False)
high_median[["Undergraduate Major", "Mid-Career Median Salary"]].head()
```
